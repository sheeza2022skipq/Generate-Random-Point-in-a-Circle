```python
import random
import matplotlib.pyplot as plt

# Number of random points
num_points = 200

# Generate random points within a specified range
x_points = [random.uniform(0, 200) for _ in range(num_points)]
y_points = [random.uniform(0, 200) for _ in range(num_points)]

# Calculate the center and radius of the circle
center_x = sum(x_points) / num_points
center_y = sum(y_points) / num_points

# Calculate the radius as the maximum distance from the center to a point
radius = max(((x - center_x)**2 + (y - center_y)**2)**0.5 for x, y in zip(x_points, y_points))

# Plot the random points
plt.scatter(x_points, y_points, label='Random Points')

# Plot the circle
circle = plt.Circle((center_x, center_y), radius, fill=False, color='red', label='Circle')
plt.gca().add_patch(circle)

# Set axis limits to ensure the entire circle is visible
plt.xlim(center_x - radius - 1, center_x + radius + 1)
plt.ylim(center_y - radius - 1, center_y + radius + 1)

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Points and Circle')
plt.legend()

# Show the plot
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
import random
import math
import matplotlib.pyplot as plt

# Number of random points
num_points = 100

# Radius of the circle
circle_radius = 5.0

# Generate random points within the circle
random_points = []
for _ in range(num_points):
    # Generate random angle (in radians)
    angle = random.uniform(0, 2 * math.pi)
    
    # Generate random radius (distance from the center)
    radius = circle_radius * math.sqrt(random.uniform(0, 1))
    
    # Convert polar coordinates to Cartesian coordinates
    x = radius * math.cos(angle)
    y = radius * math.sin(angle)
    
    random_points.append((x, y))

# Plot the random points
x_points, y_points = zip(*random_points)
plt.scatter(x_points, y_points, label='Random Points')

# Plot the circle
circle = plt.Circle((0, 0), circle_radius, fill=False, color='red', label='Circle')
plt.gca().add_patch(circle)

# Set axis limits to ensure the entire circle is visible
plt.xlim(-circle_radius - 1, circle_radius + 1)
plt.ylim(-circle_radius - 1, circle_radius + 1)

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Points within Circle')
plt.legend()

# Show the plot
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
